## Result Data Structures

Now we want to load the result files, one for each of impression and full reports, to see how well we were able to predict Pulmonary Embolism using either the impressions alone (as Chapman did) or the full reports. For these results we used [scikit-learn count vectorizer](http://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html) to perform text tokenization and preprocessing, and then used [scikit-learn ensemble trees](http://scikit-learn.org/stable/modules/generated/sklearn.ensemble.ExtraTreesClassifier.html#sklearn.ensemble.ExtraTreesClassifier) to do the prediction. The full script for the analysis [is available](https://github.com/radinformatics/pe-predictive/blob/master/classifiers/2.countVectorizer.py), and dependencies are handled via running it in its [Docker image](https://hub.docker.com/r/vanessa/pe-predictive/)

```python
# Let's load the impression results and report results - we want to visualize confusion matrices for each!
impression_confusions = pickle.load(open('impression_tree_results.pkl','rb'))
report_confusions = pickle.load(open('report_tree_results.pkl','rb'))

# What keys do we have to look at?
impression_confusions.keys()
```

```python
# The keys above indicate the batch (2,3,4) that was used for holdout (meaning test) or training.
# Batch 1 was not used, as Yu indicated it was not independent of Batch 2. Within a batch, what do we have to look at?
impression_confusions['holdout(3)-train(2|4)'].keys()
```

## Impression Classification Results

Let's iterate through all impression results, and look at scores, confusion matrices, and data frames that include:

- the report with the error
- the actual label
- the predicted label

this should give us a sense of what our classifier was having trouble with. Does it seem like something that is understandable, or could we do better? Note: I've noticed that "probably positive" is not included in any of the results, and I looked into this - the confusion only includes labels that are part of the test set, and only batch 4 has "probably positive" (although it's missing "probably negative" - this suggests that we should talk more about the batches, and how to best do (eventually) some form of cross validation.

```python
for uid,result in impression_confusions.items():
    print("\n\n### %s -----------------------------------------------\n" %(uid))
    print("Accuracy: %s" %(result["score"]))
```

Let's look at the specific errors and confusion matrix for the best performer. Note that we would probably want to look at the worst if we were writing a paper, but this is ok as we are building the model.

What I notice is that we did better than the metric reports, because a lot of "definitely negative" were confused with "probably negative," and these labels would lead to the same outcome! If we add these to the total correct, we would have:

```python
correct=434+7
incorrect=25+1+6+1
total=6+427+25+7+1+7+1
assert(correct+incorrect==total)
correct/total
```

```python
print("\n\nCONFUSION MATRIX: best performing holdout for impression")
impression_confusions['holdout(2)-train(3|4)']['confusion']
```
